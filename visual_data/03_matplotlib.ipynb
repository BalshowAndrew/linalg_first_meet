```python
import numpy as np
import matplotlib.pyplot as plt
```

# Граничные значения у координатных осей

## Метод **.set**(xlim=(xmin, xmax), ylim=(ymin, ymax))

```python
# Граничные значения осей определяются автоматически в зависимости от размерности данных:
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

plt.show()
```

```python
# Определим граничные значения осей при помощи метода .set():
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

ax.set(xlim=(-5, 30), ylim=(-1, 6))

plt.show()
```

##  Методы **.set_xlim()** и **.set_ylim()**

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

ax.set_xlim(-5, 30)
ax.set_ylim(-1, 6)

plt.show()
```

Можно отдельно отображать правую и левую границы координатных осей:

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

ax.set_xlim(xmin=-5, xmax=10) # левая граница задана равной -5, а правая - 10
ax.set_ylim(ymax=6) # правая граница задана равной 6, а левая отображается автоматически

plt.show()
```

## Функции **.xlim()** и **.ylim()**

Применяются к активной оси, если в окне только один график

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

plt.xlim(-1, 20)
plt.ylim(-1, 6)

plt.show()
```

# Уаправление метками по координатным осям

```python
from matplotlib.ticker import NullLocator, LinearLocator, MultipleLocator, IndexLocator, FixedLocator, LogLocator, MaxNLocator
```

- **set_major_locator()** - управление рисками крупной сетки
- **set_minor_locator()** - управление рисками мелкой сетки

В качестве аргументов этих функций выступает класс **matplotlib.ticker.Locator**

## Класс **NullLocator**

Класс используется для сокрытия меток по выбранной оси

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

# Создаем экземпляр класса:
lc = NullLocator()

ax.grid()
# Скрываем метки по оси x
ax.xaxis.set_major_locator(lc)


plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

# Создаем экземпляр класса:
lc = NullLocator()

ax.grid()
# скрываем метки по оси y
ax.yaxis.set_major_locator(lc)

plt.show()
```

## Класс **LinearLocator**

- Используется для задания нужного числа меток по выбранной оси графика
- Применяется редко из-за того, что часто числа получаются дробные, и работать с ними не удобно

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

ax.grid()
# Задаем по оси y только 5 меток
ax.yaxis.set_major_locator(LinearLocator(5))

plt.show()
```

## Класс **MultipleLocator**

- Позволяет задавать шаг изменения значений между рисками
- Параметр **base** определяет шаг следования значений
- Отсчет рисок производится от нуля

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

ax.grid()
# Задаем по оси y только 5 меток
ax.xaxis.set_major_locator(MultipleLocator(base=1))

plt.show()
```

## Класс **IndexLocator**

- Позволяет задавать шаг изменения значений между рисками
- Параметр **base** определяет шаг следования значений
- Отсчет рисок производится не от нуля, а от значения, заданного параметром **offset**, которое указывает относительное смещение наименьшего значения

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
# Задаем по оси y только 5 меток
ax.xaxis.set_major_locator(IndexLocator(base=0.5, offset=0))

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
# Задаем по оси y только 5 меток
ax.xaxis.set_major_locator(IndexLocator(base=0.5, offset=0.57))

plt.show()
```

## Класс **FixedLocator**

- Позволяет расставлять метки в указанных значениях
- В экземпляр класса передаем список значений тех рисок, которые мы хотим увидеть на оси

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
# Задаем по оси y только 5 меток
ax.xaxis.set_major_locator(FixedLocator([-2, -1, 0, 1, 2, 3]))

plt.show()
```

## Класс **LogLocator**

Позволяет формировать логарифмические деления по осям

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
# Задаем по оси y только 5 меток
ax.xaxis.set_major_locator(LogLocator(base=2))

plt.show()
```

## Класс **MaxNLocator**

- Прозводит разбивку по нужной координате на указанное число рисок
- В качестве аргумента задается максимальное количество рисок
- Но не обязательно, что количество рисок буте таким
- Программа сама выберет количетво рисок с учето удобныз для использования значений этих рисок

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
# Задаем по оси y только 5 меток
ax.xaxis.set_major_locator(MaxNLocator(5))

plt.show()
```

## Применение классов локаторов к минорной сетке

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

# Включаем минорную сетку:
ax.minorticks_on()
ax.grid(which='major', lw = 2)
ax.grid(which='minor')

# Отключим минорную сетку по оси x
ax.xaxis.set_minor_locator(NullLocator())

plt.show()
```
