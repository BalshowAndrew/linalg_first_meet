```python
import numpy as np
import matplotlib.pyplot as plt
```

# Гистограммы

## Гисторгамма для случайных величин

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
ax.hist(y)
ax.grid()

plt.show()
```

Количество интервалов рассчитывается автоматически

### Изменение количества интервалов

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
ax.hist(y, bins=50)
ax.grid()

plt.show()
```

# Столбчатые диаграммы

## Постоение гистограммы при помощи функции **bar()**

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

x = [f'H{i + 1}' for i in range(10)]
y = np.random.randint(1, 5, len(x))

ax.bar(x, y)
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
x = np.linspace(np.min(y), np.max(y), 10)
bars = [len(y[np.bitwise_and(y >= x[i], y < x[i + 1])]) for i in range(len(x) -1)]

ax.bar(range(len(x) - 1), bars)
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
x = np.linspace(np.min(y), np.max(y), 10)
bars = [len(y[np.bitwise_and(y >= x[i], y < x[i + 1])]) for i in range(len(x) -1)]

ax.barh(range(len(x) - 1), bars)
ax.grid()

plt.show()
```

## Параметры функции **bar()**

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

x = [f'H{i + 1}' for i in range(10)]
y = np.random.randint(-20, 20, len(x))

ax.bar(x, y, width=0.5, linewidth=2, edgecolor='r', yerr=2, bottom=10)
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

x = np.arange(10)
y1 = np.random.randint(3, 20, len(x))
y2 = np.random.randint(3, 20, len(x))
w = 0.3
ax.bar(x - w/2, y1, width=w)
ax.bar(x + w/2, y2, width=w)

ax.grid()

plt.show()
```

# Круговые даиграммы

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

vals = [10, 40, 23, 30, 7]
labels = ['Toyota', 'BMW', 'Lexus', 'Audi', 'Lada']
ax.pie(vals, labels=labels)

ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

vals = [10, 40, 23, 30, 7]
labels = ['Toyota', 'BMW', 'Lexus', 'Audi', 'Lada']
exp = (0.1, 0.2, 0, 0, 0)
ax.pie(vals, labels=labels, autopct='%.2f', explode=exp, shadow=True)

ax.grid()

plt.show()
```
