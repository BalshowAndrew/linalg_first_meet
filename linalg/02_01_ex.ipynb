# Векторы. Часть 1. Упражнения

## Упражнение 2.1

```python
v = np.array([1, 2])
w = np.array([4, -6])
a = v + w
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

a1 = plt.arrow(0, 0, v[0], v[1], color='blue', width=0.1, length_includes_head=True)
a2 = plt.arrow(v[0], v[1], w[0], w[1], color='red', width=0.1, length_includes_head=True)
a3 = plt.arrow(0, 0, a[0], a[1], color='green', width=0.1, length_includes_head=True)


plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend((a1, a2, a3), ['v', 'w', 'v + w'])
plt.grid()
```

```python
v = np.array([1, 2])
w = np.array([4, -6])
a = v - w
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

a1 = plt.arrow(0, 0, v[0], v[1], color='blue', width=0.1, length_includes_head=True)
a2 = plt.arrow(0, 0, w[0], w[1], color='red', width=0.1, length_includes_head=True)
a3 = plt.arrow(w[0], w[1], a[0], a[1], color='green', width=0.1, length_includes_head=True)


plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend((a1, a2, a3), ['v', 'w', 'v - w'])
plt.grid()
```

## Упражнение 2.2

```python
def norm(v):
    return np.sqrt(np.sum(v**2))
```

```python
def v_norm(n):
    for i in range(n):
        v = np.array([np.random.randint(10) for i in range (np.random.randint(10))])
        print(v, np.linalg.norm(v) == norm(v))
```

```python
v_norm(5)
```

## Упражнение 2.3

```python
def unit_v(w):
    return (1 / np.linalg.norm(w)) * w
```

```python
w = np.array([133, -5, 7])
```

```python
unit_w = unit_v(w)
print(unit_w, np.linalg.norm(unit_w))
```

```python
w = np.zeros((4, 1))
unit_w = unit_v(w)
print(w, np.linalg.norm(w))
print()
print(unit_w, np.linalg.norm(unit_w))
```

## Упражнение 2.4

```python
def changeMod(v, n):
    uniqV = v / np.linalg.norm(v)
    return n * uniqV
```

```python
v = np.array([-3, 2])
v_2 = changeMod(v, -2)
v_1 = v / np.linalg.norm(v)
print(v)
print(v_1)
print(v_2)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

a1 = plt.arrow(0, 0, v[0], v[1], color='blue', width=.1, length_includes_head=True, alpha=.5)
a2 = plt.arrow(0, 0, v_1[0], v_1[1], color='red', width=.1, length_includes_head=True, alpha=.5)
a3 = plt.arrow(0, 0, v_2[0], v_2[1], color='green', width=.1, length_includes_head=True, alpha=.5)


plt.axis('square')
plt.axis([-4, 4, -4, 4])
plt.legend((a1, a2, a3), ['v', 'v_1', 'v_2'])
plt.grid()
```

```python
v = np.array([2, 0, 0])
v_2 = changeMod(v, 4)
v_1 = v / np.linalg.norm(v)
print(v)
print(v_1)
print(v_2)
```
