```python
import numpy as np
import matplotlib.pyplot as plt
```

## Задаем координаты фукции **plot**

```python
# создаем координату y а значения координаты x будут формироваться из индексов координаты y
y = np.array([1, 3, -6, 0, 4])
plt.plot(y)
plt.show()
```

```python
# зададим обе координаты:
x = np.array([4, 5, 6, 7, 8])
y = np.array([1, 3, -6, 0, 4])
plt.plot(x, y)
plt.show()
```

### Строим квадрат

```python
x = np.array([1, 1, 5, 5, 1])
y = np.array([1, 5, 5, 1, 1])
plt.plot(x, y)
plt.show()
```

### График с сеткой

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])

plt.plot(x, y)
plt.grid()
plt.show()
```

## Размещаем два графика на одной плоскости

```python
# Первый вариант:
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

plt.plot(x, y, x2, y2)
plt.grid()
plt.show()
```

```python
# Второй вариант:
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

plt.plot(x, y)
plt.plot(x2, y2)
plt.grid()
plt.show()
```

## Оформление графиков

### Типы линий графика

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])

plt.plot(x, y, '--') # штриховая линия графика
plt.grid()
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])

plt.plot(x, y, '-.') # штрих-пунктирная линия графика
plt.grid()
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])

plt.plot(x, y, ':') # пунктирная линия графика
plt.grid()
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])

plt.plot(x, y, ' ') # Без рисования линии
plt.grid()
plt.show()
```

### Объект **Line2D**

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])

lines = plt.plot(x, y)
plt.grid()
print(lines)
plt.show()
```

Был создан объект **lines** - это экземпляр класса **Line2D**, через него мы можем управлять графиком

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])

lines = plt.plot(x, y)
plt.grid()
plt.setp(lines, linestyle='-.')
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

lines = plt.plot(x, y, x2, y2)
plt.setp(lines, linestyle='-.')

plt.grid()
plt.show()
```

### Изменим цвет графиков

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

lines = plt.plot(x, y, '--r', x2, y2, ':m') # цвет для каждого графика отдельно

plt.grid()
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

lines = plt.plot(x, y, x2, y2, color='g') # назначаем один цвет для всех графиков одновременно

plt.grid()
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

lines = plt.plot(x, y, x2, y2, color=(0, 0, 1, 0.6)) # кодируем цвет (r, g, b, alpha-канал)

plt.grid()
plt.show()
```

### Меняем маркеры у точек

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

lines = plt.plot(x, y, 'r--o', x2, y2, 'm:s') # цвет для каждого графика отдельно

plt.grid()
plt.show()
```

Перечень маркеров смотри в документации

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

lines = plt.plot(x, y, x2, y2, color=(0, 0, 1, 0.6), marker='s') # одинаковый маркер для всех графиков

plt.grid()
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

lines = plt.plot(x, y, x2, y2, color=(0, 0, 1, 0.6), marker='s', markerfacecolor='w') # цвет для маркера

plt.grid()
plt.show()
```

### Использование функции **setp**

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])

lines = plt.plot(x, y, x2, y2)
plt.setp(lines, linestyle='-.', color=(0, 0, 1), marker='s', markerfacecolor='black', linewidth=3)

plt.grid()
plt.show()
```

### Заливка областей графика

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y)

plt.grid()
plt.show()
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y, 0.5)

plt.grid()
plt.show()
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y, 0.5, where=(y < 0))

plt.grid()
plt.show()
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y, where=(y < 0), color='r', alpha=0.5)
plt.fill_between(x, y, where=(y > 0), color='g', alpha=0.5)

plt.grid()
plt.show()
```
