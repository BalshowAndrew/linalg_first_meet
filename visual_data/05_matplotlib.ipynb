```python
import numpy as np
import matplotlib.pyplot as plt
```

# Размещение стандартных текстовых элементов на графике

## Отображение мажорной сетки

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])
lines = plt.plot(x, y, x2, y2)
plt.grid()
plt.show()
```

## Отображение минорной сетки

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = np.array([0, 1, 2, 3])
x2 = np.array([i + 1 for i in y2])
lines = plt.plot(x, y, x2, y2)

plt.minorticks_on()
plt.grid(which='major', color='#444', linewidth = 0.7)
plt.grid(which='minor', color='#aaa', ls=':')
plt.show()
```

**NB:** Мелкая сетка снижает производительность при отображении графиков, поэтуму ее без особой необходимости лучше не использовать.

## Размещение текстовых элементов

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
plt.figtext(0.2, 0.8, 'Текст в области окна')
fig.suptitle('Заголовок окна')
ax.set_xlabel('Ox')
ax.set_ylabel('Oy')
ax.text(0.05, 0.1, 'Произвольный текст в координатных осях')
ax.annotate('Аннотация', xy=(0.2, 0.4), xytext=(0.5, 0.6),
           arrowprops={'facecolor': 'gray', 'shrink': 0.1})
plt.show()
```

## Определяем размер шрифта

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
plt.figtext(0.2, 0.8, 'Текст в области окна', fontsize=18)
fig.suptitle('Заголовок окна', fontsize=24)
ax.set_xlabel('Ox')
ax.set_ylabel('Oy')
ax.text(0.05, 0.1, 'Произвольный текст в координатных осях', fontsize=14)
ax.annotate('Аннотация', xy=(0.2, 0.4), xytext=(0.5, 0.6),
           arrowprops={'facecolor': 'gray', 'shrink': 0.1}, fontsize=16)
plt.show()
```

## Задаем цвет шрифта

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
plt.figtext(0.2, 0.8, 'Текст в области окна', fontsize=18, color='green')
fig.suptitle('Заголовок окна', fontsize=24, color='blue')
ax.set_xlabel('Ox')
ax.set_ylabel('Oy')
ax.text(0.05, 0.1, 'Произвольный текст в координатных осях', fontsize=14, color='gray')
ax.annotate('Аннотация', xy=(0.2, 0.4), xytext=(0.5, 0.6),
           arrowprops={'facecolor': 'gray', 'shrink': 0.1}, fontsize=16, color='red')
plt.show()
```

## Цветовая заливка областей и использование параметра **boxstyle**

```python
fig = plt.figure(figsize=(7, 4))
fig.set(facecolor='#eee') # цвет заливки окна
ax = fig.add_subplot()
ax.set(facecolor='#AAFFAA') # цвет заливки координатной оси
plt.figtext(0.2, 0.8, 'Текст в области окна', fontsize=18, color='green')
fig.suptitle('Заголовок окна', fontsize=24, color='blue')
ax.set_xlabel('Ox')
ax.set_ylabel('Oy')
ax.text(0.05, 0.1, 'Произвольный текст в координатных осях', fontsize=14,
       bbox={'boxstyle': 'darrow', 'facecolor': '#AAAAFF'})
ax.annotate('Аннотация', xy=(0.2, 0.4), xytext=(0.5, 0.6),
           arrowprops={'facecolor': 'gray', 'shrink': 0.1}, fontsize=16, color='red')
plt.show()
```
