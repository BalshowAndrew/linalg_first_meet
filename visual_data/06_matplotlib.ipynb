```python
import numpy as np
import matplotlib.pyplot as plt
```

# Добавляем легенду на график

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(0, 5, 0.25), '--o', label='line 1')
ax.plot(np.arange(0, 10, 0.5), ':s',label='line 2')
ax.legend()
plt.show()
```

## Задаем названия графиков в легенде

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(0, 5, 0.25), '--o', label='line 1')
ax.plot(np.arange(0, 10, 0.5), ':s',label='line 2')
ax.legend(['График 1', 'График 2'])
plt.show()
```

## Изменим порядок следования графиков в легенде

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s')
ax.legend((line2, line1), ['График 2', 'График 1'])
plt.show()
```

## Расположение легенды

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s')
# Разместим легенду вверху справа:
ax.legend((line2, line1), ['График 2', 'График 1'], loc='upper right')
plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s')
# Разместим легенду внизу по центру:
ax.legend((line2, line1), ['График 2', 'График 1'], loc='lower center')
plt.show()
```

Укажем более точное расположение легенды с помощью параметра **bbox_to_anchor=()**:

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s')
# Разместим легенду внизу по центру:
ax.legend((line2, line1), ['График 2', 'График 1'], bbox_to_anchor=(0.4, 0.7))
plt.show()
```

Matplotlib позволяет записывать формулы в формате **tex**:

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s')
# Разместим легенду внизу по центру:
ax.legend((line2, line1), [r'$f(x) = a \cdot b + c$', r'$f(x) = k \cdot x + b$'])
plt.show()
```

## Оформление элементов в окне легенды

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s')
# Разместим легенду внизу по центру:
ax.legend((line2, line1), [r'$f(x) = a \cdot b + c$', r'$f(x) = k \cdot x + b$'],
         facecolor='#aaa', framealpha=0.5)
plt.show()
```

# Рисование произвольных фигур в координатных осях

## Рисование ломаной линии

```python
# импортируем класс Line2D
from matplotlib.lines import Line2D
```

```python
# Задаем координаты ломаной линии:
l1 = Line2D([1, 2, 3], [1, 2, 4])

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s')

# Добавляем ломаную линию в координатную ось:
ax.add_line(l1)

plt.show()
```

Для того, чтобы ломаная линия изображалась на графике, надо задать нужные пределы осей:

```python
# Задаем координаты ломаной линии:
l1 = Line2D([1, 2, 3], [1, 2, 4])

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
# Добавляем ломаную линию в координатную ось:
ax.add_line(l1)
# Задаем нужные пределы осей:
ax.set(xlim=(1, 3), ylim=(1, 4))

plt.show()
```

```python
# Задаем ломаную косинусоиду:
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
cos = Line2D(x, np.cos(x))

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
# Добавляем ломаную линию в координатную ось:
ax.add_line(cos)
# Задаем нужные пределы осей:
ax.set(xlim=(-2*np.pi, 2*np.pi), ylim=(-1, 1))

plt.show()
```

## Другие классы геометрических фигур

```python
# Импортируем все классы:
from matplotlib.patches import *
```

### Создадим прямоугольник при помощи класса **Rectangle**

```python
rect = Rectangle((0, 0), 2.5, 0.5, facecolor='g')

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
# Добавляем ломаную линию в координатную ось:
ax.add_patch(rect)
# Задаем нужные пределы осей:
ax.set(xlim=(-2*np.pi, 2*np.pi), ylim=(-1, 1))

plt.show()
```
