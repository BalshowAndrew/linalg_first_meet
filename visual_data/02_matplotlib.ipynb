## Функция **subplots()**

Функция используется для создания нескольких координатных осей в одной фигуре

### Создание координатрых осей

```python
# Делаем разбивку на две строки и два столбца:
f, ax = plt.subplots(2, 2)

plt.show()
```

Функция **subplots()** возвращает фигуру: **f** и список координатных осей: **ax**

```python
# Список координатных осей:
ax
```

### Размещение на координатных осях графиков

```python
f, ax = plt.subplots(2, 2)

# работаем с первой координатной осью:
ax[0,0].plot(np.arange(0, 5, 0.2))
ax[0,0].grid()
# работаем со второй координатной осью:
ax[0,1].plot(np.arange(5, 0, -0.2))
ax[0,1].grid()

plt.show()
```

### Работа с фигурой

```python
f, ax = plt.subplots(2, 2)

# задаем размер фигуры 7 х 4 в дюймах:
f.set_size_inches(7, 4)
# задаем серый цвет фона:
f.set_facecolor('#eee')

ax[0,0].plot(np.arange(0, 5, 0.2))
ax[0,0].grid()

ax[0,1].plot(np.arange(5, 0, -0.2))
ax[0,1].grid()

plt.show()
```

## Создание нескольких фигур

```python
# Первая фигура:
fig = plt.figure(figsize=(7, 4))
plt.plot(np.arange(0, 5, 0.2))

# Вторая фигура:
f, ax = plt.subplots(2, 2)
f.set_size_inches(7, 4)
f.set_facecolor('#eee')
ax[0,0].plot(np.arange(0, 5, 0.2))
ax[0,0].grid()
ax[0,1].plot(np.arange(5, 0, -0.2))
ax[0,1].grid()

plt.show()
```

### Метод **.add_axes()**

```python
# Создадим отдельно координатную ось для первой фигуры:
fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_axes([0.0, 0, 1.0, 1.0])
# и вэтой оси разместим график:
ax1.plot(np.arange(0, 5, 0.2))

# Вторая фигура:
f, ax = plt.subplots(2, 2)
f.set_size_inches(7, 4)
f.set_facecolor('#eee')
ax[0,0].plot(np.arange(0, 5, 0.2))
ax[0,0].grid()
ax[0,1].plot(np.arange(5, 0, -0.2))
ax[0,1].grid()

plt.show()
```

### Метод **.add_subplot()**

```python
# создаем фигуру и в ней координатную ось:
fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot(1, 3, 1)
# и вэтой оси разместим график:
ax1.plot(np.arange(0, 5, 0.2))
plt.show()
```

## Компановка графиков с помощью **GridSpec**

- Импорт модуля **matplotlib.gridspec**
- Создаем фигуру - **fig**
- Создание экземпляра класса **GridSpec**(ncols=3, nrows=2, figure=**fig**)
- Для создания координатных осей можно использовать следующие функции:
    - *plt.subplot()*;
    - *fig.add_subplot()*;
- В качестве аргумента используем экземпляр класса **GridSpec** в котором указываем, где будем размещать наши координатные оси
- Для этого используется синтаксис среза NumPy

```python
from matplotlib.gridspec import GridSpec
```

```python
fig  = plt.figure(figsize=(7, 4))
gs = GridSpec(ncols=3, nrows=2, figure=fig)

ax1 = plt.subplot(gs[0, 0])
ax1.plot(np.arange(0, 5, 0.2))
ax2 = fig.add_subplot(gs[1, 0:2])
ax2.plot(np.random.random(10))
ax3 = fig.add_subplot(gs[:, 2])
ax3.plot(np.random.random(10))

plt.show()
```

### Указание размеров координатных осей:

```python
# Списки с указанием доли каждой строки и столбца:
ws = [1, 2, 5]
hs = [2, 0.5]

# Передадим эти списки в параметры экземпляра класса
fig  = plt.figure(figsize=(7, 4))
gs = GridSpec(ncols=3, nrows=2, figure=fig, width_ratios=ws, height_ratios=hs)

ax1 = plt.subplot(gs[0, 0])
ax1.plot(np.arange(0, 5, 0.2))
ax2 = fig.add_subplot(gs[1, 0:2])
ax2.plot(np.random.random(10))
ax3 = fig.add_subplot(gs[:, 2])
ax3.plot(np.random.random(10))

plt.show()
```
