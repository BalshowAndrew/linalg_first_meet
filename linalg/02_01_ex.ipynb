```python
import numpy as np
import matplotlib.pyplot as plt
```

# Векторы. Часть 1. Упражнения

## Упражнение 2.1

```python
v = np.array([1, 2])
w = np.array([4, -6])
a = v + w
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

a1 = plt.arrow(0, 0, v[0], v[1], color='blue', width=0.1, length_includes_head=True)
a2 = plt.arrow(v[0], v[1], w[0], w[1], color='red', width=0.1, length_includes_head=True)
a3 = plt.arrow(0, 0, a[0], a[1], color='green', width=0.1, length_includes_head=True)


plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend((a1, a2, a3), ['v', 'w', 'v + w'])
plt.grid()
```

```python
v = np.array([1, 2])
w = np.array([4, -6])
a = v - w
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

a1 = plt.arrow(0, 0, v[0], v[1], color='blue', width=0.1, length_includes_head=True)
a2 = plt.arrow(0, 0, w[0], w[1], color='red', width=0.1, length_includes_head=True)
a3 = plt.arrow(w[0], w[1], a[0], a[1], color='green', width=0.1, length_includes_head=True)


plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend((a1, a2, a3), ['v', 'w', 'v - w'])
plt.grid()
```

## Упражнение 2.2

```python
def norm(v):
    return np.sqrt(np.sum(v**2))
```

```python
def v_norm(n):
    for i in range(n):
        v = np.array([np.random.randint(10), np.random.randint(10)])
        print(np.linalg.norm(v) == norm(v))
```

```python
v_norm(5)
```
